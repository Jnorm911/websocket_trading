```python
import pandas as pd


# Set display options to show all rows and columns
pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)
df = pd.read_csv("selected_features.csv")
filtered_df = df[df["iteration"] >= 2]
count = filtered_df[
    (filtered_df["iteration"] == 8) & (filtered_df["reoccurrence"] == 8)
]["feature"].count()
print("Total count of features with iteration = 8 and reoccurrence = 8:", count)
best_features = list(
    filtered_df[(filtered_df["iteration"] == 8) & (filtered_df["reoccurrence"] == 8)][
        "feature"
    ]
)

print(best_features)

filtered_df
```
