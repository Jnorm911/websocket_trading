```python
import pandas as pd
import yfinance as yf
import pandas_ta as ta

df = pd.DataFrame()  # Empty DataFrame

# Load data
df = df.ta.ticker("BTC-USD", period="7d", interval="1m")

# Take a peek
df.tail()
```

```python
# List of indicator categories
df.ta.categories

# Running a Categorical Strategy only requires the Category name
df.ta.strategy("Momentum")  # Default values for all Momentum indicators
df.ta.strategy("overlap")  # Override all Overlap 'length' attributes
df.tail()
```
